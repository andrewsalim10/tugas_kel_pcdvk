```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('merah_hijau.jpg')



plt.imshow( cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
b, g, r = cv2.split(img)
```

```python
plt.imshow(b)
```

```python
plt.imshow(g)
```

```python
plt.imshow(r)
```

```python
new_img = cv2.merge( [b, g, r] )

plt.imshow( cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB))
```
